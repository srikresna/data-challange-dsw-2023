<h3>Binning</h3>

```python
df['Tenure Months'].describe()
```

```python
def categorize_tenure_percentile(tenure):
    if tenure <= df['Tenure Months'].quantile(0.25):
        return "Low Tenure"
    elif df['Tenure Months'].quantile(0.25) < tenure <= df['Tenure Months'].quantile(0.75):
        return "Medium Tenure"
    else:
        return "High Tenure"

df['Tenure_Category'] = df['Tenure Months'].apply(categorize_tenure_percentile)
```

```python
df.head()
```

```python
df['CLTV (Predicted Thou. IDR)'].describe()
```

```python
def categorize_cltv_percentile(x):
    if x <= df['CLTV (Predicted Thou. IDR)'].quantile(0.25):
        return "Low CLTV"
    elif df['CLTV (Predicted Thou. IDR)'].quantile(0.25) < x <= df['CLTV (Predicted Thou. IDR)'].quantile(0.75):
        return "Medium CLTV"
    else:
        return "High CLTV"
    
df['CLTV_Category'] = df['CLTV (Predicted Thou. IDR)'].apply(categorize_cltv_percentile)
df.head()
```

```python
df['Monthly Purchase (Thou. IDR)'].describe()
```

```python
def categorize_monthlyPurchase_percentile(x):
    if x <= df['Monthly Purchase (Thou. IDR)'].quantile(0.25):
        return "Low Monthly Purchase"
    elif df['Monthly Purchase (Thou. IDR)'].quantile(0.25) < x <= df['Monthly Purchase (Thou. IDR)'].quantile(0.75):
        return "Medium Monthly Purchase"
    else:
        return "High Monthly Purchase"
    
df['MonthlyPurchase_Category'] = df['Monthly Purchase (Thou. IDR)'].apply(categorize_monthlyPurchase_percentile)
df.head()
```

```python
df['tenure_cltv_category'] = df['Tenure_Category'] + " " + df['CLTV_Category']
df['tenure_monthly_purchase_category'] = df['Tenure_Category'] + " " + df['MonthlyPurchase_Category']
df['cltv_monthly_purchase_category'] = df['CLTV_Category'] + " " + df['MonthlyPurchase_Category']
df['tenure_cltv_monthly_purchase_category'] = df['Tenure_Category'] + " " + df['CLTV_Category'] + " " + df['MonthlyPurchase_Category']

df.head()
```

```python
df['tenure_cltv_churn_rate'] = df['tenure_cltv_category'].map(df.groupby('tenure_cltv_category')['Churn Label'].mean())
df['tenure_monthly_purchase_churn_rate'] = df['tenure_monthly_purchase_category'].map(df.groupby('tenure_monthly_purchase_category')['Churn Label'].mean())
df['cltv_monthly_purchase_churn_rate'] = df['cltv_monthly_purchase_category'].map(df.groupby('cltv_monthly_purchase_category')['Churn Label'].mean())
df['tenure_cltv_monthly_purchase_churn_rate'] = df['tenure_cltv_monthly_purchase_category'].map(df.groupby('tenure_cltv_monthly_purchase_category')['Churn Label'].mean())

df.head()
```
