```python
# if no = 0, yes = 1, no internet service = 0

service = ['Games Product', 'Music Product', 'Education Product', 'Call Center', 'Video Product', 'Use MyApp']

for i in service:
    df[i] = df[i].replace(['No', 'Yes', 'No internet service'], [0, 1, 0])

df.head()
```

```python
df['service_used'] = df[service].sum(axis=1)
```

```python
# make feature churn rate by payment method
df['Churn Label'] = df['Churn Label'].replace(['Yes', 'No'], [1, 0])
df.groupby('Payment Method')['Churn Label'].mean()

df['payment_method_churn_rate'] = df['Payment Method'].map(df.groupby('Payment Method')['Churn Label'].mean())
df.head()
```

```python
# make feature churn rate by location
df.groupby('Location')['Churn Label'].mean()

df['location_churn_rate'] = df['Location'].map(df.groupby('Location')['Churn Label'].mean())
df.head()
```

```python
# make feature churn rate by device owned
df.groupby('Device Class')['Churn Label'].mean()

df['device_churn_rate'] = df['Device Class'].map(df.groupby('Device Class')['Churn Label'].mean())
df.head()
```

```python
df.groupby(['Games Product', 'Music Product', 'Education Product', 'Video Product', 'Call Center', 'Use MyApp'])['Churn Label'].mean()

df['service_churn_rate'] = df.groupby(['Games Product', 'Music Product', 'Education Product', 'Video Product', 'Call Center', 'Use MyApp'])['Churn Label'].transform('mean')

df.head()
```

```python
df.groupby(['Device Class', 'Games Product'])['Churn Label'].mean()

df['device_games_churn_rate'] = df.groupby(['Device Class', 'Games Product'])['Churn Label'].transform('mean')

df.head()
```

<h3>Binning</h3>

```python
df['Tenure Months'].describe()
```

```python
def categorize_tenure_percentile(tenure):
    if tenure <= df['Tenure Months'].quantile(0.25):
        return "Low Tenure"
    elif df['Tenure Months'].quantile(0.25) < tenure <= df['Tenure Months'].quantile(0.75):
        return "Medium Tenure"
    else:
        return "High Tenure"

df['Tenure_Category'] = df['Tenure Months'].apply(categorize_tenure_percentile)
```

```python
df['CLTV (Predicted Thou. IDR)'].describe()
```

```python
def categorize_cltv_percentile(x):
    if x <= df['CLTV (Predicted Thou. IDR)'].quantile(0.25):
        return "Low CLTV"
    elif df['CLTV (Predicted Thou. IDR)'].quantile(0.25) < x <= df['CLTV (Predicted Thou. IDR)'].quantile(0.75):
        return "Medium CLTV"
    else:
        return "High CLTV"
    
df['CLTV_Category'] = df['CLTV (Predicted Thou. IDR)'].apply(categorize_cltv_percentile)
df.head()
```

```python
def categorize_monthlyPurchase_percentile(x):
    if x <= df['Monthly Purchase (Thou. IDR)'].quantile(0.25):
        return "Low Monthly Purchase"
    elif df['Monthly Purchase (Thou. IDR)'].quantile(0.25) < x <= df['Monthly Purchase (Thou. IDR)'].quantile(0.75):
        return "Medium Monthly Purchase"
    else:
        return "High Monthly Purchase"
    
df['MonthlyPurchase_Category'] = df['Monthly Purchase (Thou. IDR)'].apply(categorize_monthlyPurchase_percentile)
df.head()
```
